```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Coeficientes de las ecuaciones
A = np.array([[2, -1, 3],
              [4, 0, -2],
              [1, 3, -1]])
b = np.array([16, -4, -9])

# Resolver el sistema de ecuaciones
solucion = np.linalg.solve(A, b)
x1, x2, x3 = solucion

# Crear una malla de puntos en los planos XY
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

# Calcular los valores de z para cada plano
Z1 = (16 - 2*X + Y) / 3
Z2 = (4*X + 2) / 2
Z3 = (-9 - X + 3*Y)

# Graficar los planos
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z1, alpha=0.5, rstride=100, cstride=100, color='r')
ax.plot_surface(X, Y, Z2, alpha=0.5, rstride=100, cstride=100, color='g')
ax.plot_surface(X, Y, Z3, alpha=0.5, facecolors='b', rstride=100, cstride=100)

# Graficar el punto de intersección
ax.scatter(x1, x2, x3, color='k', s=100, label='Punto de intersección')

# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar la gráfica
plt.title('Planos en 3D')
plt.legend()
plt.show()


```
